```python
# import modules
%matplotlib inline
from IPython import display

import os
import time
import pylab
import random

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.misc import imread, imresize

project_root = os.path.join(os.path.expanduser('~'), 
                            'workspace/.project/right_whale')
data_root = os.path.join(project_root, 'data')
subs_root = os.path.join(project_root, 'subs')
```

```python
# load train and test 
train = pd.read_csv(os.path.join(data_root, 'train.csv'))
test = pd.read_csv(os.path.join(data_root, 'sample_submission.csv'))

train_annotations = pd.read_csv(os.path.join(data_root, 'train_with_annotations.csv'))
```

```python
# visualize data
img = imread(data_root+"/traincrops/"+train.Image[0])
pylab.imshow(img)
```

```python
# visualize data
pylab.rcParams['figure.figsize'] = 16, 12

for _ in range(100):
    i = random.randrange(0, train.shape[0])

    display.display(display.Image(data_root+"/train/"+train.Image[i]))
    
    time.sleep(5)
    
    display.clear_output()

    ax = plt.gca()
    img = imread(data_root+"/train/"+train.Image[i])
    ax.imshow(img)    
       
    x = train_annotations.ix[i, 2]
    y = train_annotations.ix[i, 3]
    w = train_annotations.ix[i, 4]
    h = train_annotations.ix[i, 5]
    ax.add_patch(patches.Rectangle((x, y), w, h, fill = False, color = 'red'))
    
    ax.text(x - 10, y - 10, 'mike testing', color = 'blue', bbox = {'facecolor':'white'})

    pylab.axis('off')
    plt.show()
    
    time.sleep(5)
    display.clear_output()
```
